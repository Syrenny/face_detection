```python
import matplotlib.image as mpimg
from matplotlib import pyplot as plt
import matplotlib.image as mpimg
%matplotlib inline

def plot_img(img: np.array, figsize = (20, 10), savefig=None):
    """ 
    img -> np.array,
    figsize -> tuple,
    savefig -> str
    """
    plt.figure(figsize = figsize)
    plt.imshow(img)
    if savefig:
        plt.savefig(savefig)
    plt.show()
```
